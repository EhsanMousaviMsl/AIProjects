### Illostration of Taylor expansion of sin(x):

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.special import factorial
```

```python
x = np.linspace(-10, 10, 2000)
y = np.sin(x)
plt.plot(x, y, 'k', linewidth=2)
plt.xlim(-10, 10)
plt.ylim(-10, 10)


# first order tylor expansion

P = [0, 1] # x + 0
yT1 = np.polyval(P, x)
plt.plot(x, yT1, 'b--', linewidth= 1.2)

# third order tylor expansion
P = [-1/factorial(3), 0, 1, 0] # -(1/3!)x^3 + 0(x^2) + x + 0 
yT3 = np.polyval(P, x)
plt.plot(x, yT3, 'r--', linewidth= 1.2)

# fifth order tylor expansion

P = [1/factorial(5), 0, -1/factorial(3),0, 1, 0]
yT5 = np.polyval(P, x)
plt.plot(x, yT5, 'g--', linewidth= 1.2)


plt.legend(['sin(x)', '1st order', '3rd order', '5th order', '7th order', '9th order'])
plt.show()
```
