```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
t = np.linspace(0, 10, 1000)
x0 = [2 , -3]
A = np.array([
    [0, 1],
    [-2, -3]
])

x = np.exp(-t) + np.exp(-2 * t)


def linear_ode(t, x):
    return A @ x

linear_ode_solution = solve_ivp(linear_ode, (0, 10), x0, t_eval=t)
xRungKutta = linear_ode_solution.y

plt.plot(t, x, 'r')
plt.plot(t, xRungKutta[0, :], 'b--')
plt.xlabel('time')
plt.ylabel('position')
plt.legend(['exact solution', 'RK4'])
```
