```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def weatherModel(today, day):
    x_today = [0, 0, 0]  # rainy, nice, cloudy  
    if today == 'rainy':
        x_today[0] = 1
    elif today == 'nice':
        x_today[1] = 1
    elif today == 'cloudy':
        x_today[2] = 1
    else:
        print("Error!")
        return
    A = np.array([[0.5, 0.5, 0.25], [0.25, 0, 0.25], [0.25, 0.5, 0.5]])
    forecast = np.zeros((day, 3))
    for i in range(day):
        x_tommorow = A @ x_today
        print(f'day {i + 1}:')
        print(f'Rainy: {x_tommorow[0]:.4f}, Nice: {x_tommorow[1]:.4f}, Cloudy: {x_tommorow[2]:.4f}')
        x_today = x_tommorow
        forecast[i, :] = x_tommorow

    return forecast


```

```python
day = 10
forecast = weatherModel('rainy',day)

```

```python
plt.plot(forecast)
plt.grid()
plt.show()
```
